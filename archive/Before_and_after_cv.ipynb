# Pilot Study: Distribution of the best performing model before and after CV

```r
library(dplyr)
library(tibble)
library(readr)
library(ggplot2)
library(purrr)
library(tidyr)
```

```r
dir = "~/Project/Alz/"
Tissue = c("DLPFC","AC","PCC")
Proj = c("Alz_DLPFC_SNP","Alz_AC_SNP","Alz_PCC_SNP")
sur_pre_cv = "/result_Arch/"
sur_po_cv = "/result/"
tb_pre_cv = tibble(Tissue,Proj)%>%mutate(path = map(Proj,~paste(c(dir,.x,sur_pre_cv,"all_result.txt"),collapse  = "") ),
                                       data = map(path,~read_delim(.x,delim = "\t")%>%filter(is.na(PANEL) )) 
                                        )
tb_po_cv = tibble(Tissue,Proj)%>%mutate(path = map(Proj,~paste(c(dir,.x,sur_po_cv,"all_result.txt"),collapse  = "") ),
                                       data = map(path,~read_delim(.x,delim = "\t")%>%filter(is.na(PANEL) )) 
                                       )

```

```r
tb_po_cv%>%unnest(data)%>%ggplot(aes(x = MODEL))+geom_bar()
```

```r
tb_pre_cv%>%unnest(data)%>%write_delim("~/pre_cv_fusion_result.txt","\t")
tb_po_cv%>%unnest(data)%>%write_delim("~/po_cv_fusion_result.txt","\t")
```
